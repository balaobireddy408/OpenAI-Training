```python
# whenever we are selecting features, we want features that are
#    1. Independency between columns themseleves
#    2. Whether the column is required for OUTPUT! [Diagnosis]
# Correlation
#   1. Directly proportional (Positive Correlation)
#   2. Inversely proportional (Negative Correlation)
#   3. ZERO Correlation

# Correlation can only happen between numbers!
all_col_names = list(df.columns)[:16] # we are selecting top 16 only to make our visualizations
# easier to look at!

all_col_names.remove('diagnosis')

numbers_only = df.loc[:, all_col_names]


cors = numbers_only.corr()
cors
```

```python
# correlations can be best visualized with a HEATMAP
import seaborn as sns
sns.heatmap(cors, cmap = 'coolwarm')

# ID -> entire row and the column is tending to 0!

# cell are sphrical -> image of it-> 2-D (Circle)
# area of a circle = pi * r * r
# perimeter of a circle = 2 * pi * r

# radius -> in presence of a mathematical formulae, you will see tending to 1 correlation

# this means-> we dont need all 3 of them-> but could have worked with ANY 1 of them!
# (radius, perimeter and area)

# dy/dx -> for a tiny change (delta) in x, what is the RATE of change of Y
# between -1 to 1
# rate of change-> correlation
# absolute value-> covariance
```

```python
# scatter plot-> distribution of inputs with respect to the output
encoding_values = {'M': 1, 'B':0}
encoding_logic = lambda val: encoding_values[val]
y = data['diagnosis'].map(encoding_logic)

sm = pd.plotting.scatter_matrix(numbers_only, c=y, figsize=(20,20))

# ID column-> observe that your covariance w.r.t the output
# is parallel to X axis and Y axis!
# dy/d_id -> 0

# we want to keep all POSITIVE and all NEGATIVE
# we want to remove all 0 correlations!
```

```python
#conclusion -> ID should NOT participant in ML!
# 2 graphs-> correlation heatmap, covariance scatter plot-> both have said the same thing
# changing ID does not do anything to input or to output!!!

# ID-> but shouldn't be eliminated either! we will lose the file no./patient no. and not know
# which ID belongs to which patient
# simplest way forward-> ID has all unique value

data['id'].value_counts().count() # we already know that there are 569 rows in total

# since ID is a unique column-> we can use it as our INDEX instead!
```

```python
data = data.iloc[:, :-1]
df = data.set_index('id')
```

```python
df.head()
```

```python
y = df['diagnosis']
x = df.iloc[:, 1:]
x.head()
```

```python
# SPLIT DATA-> training and scoring (testing is also known as SCORING)
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.tail()
```

```python
ytest.tail()
```

```python
# remember -> ML happens only on numbers!
encoding = {'M':1, 'B':0}
logic = lambda val: encoding[val]
y_train_encoded = ytrain.apply(logic)
y_test_encoded = ytest.apply(logic)
```

```python
# Machine Learning -> TRAINING the MODEL -> FITTING the model
# Fitting-> trying to fit a boundary into my dataset!

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(xtrain, y_train_encoded)

```

```python
predictions = model.predict(xtest)
```

```python
predictions[:5]
```

```python
predictions.min(), predictions.max()
```

```python
# calculate a threshold between min and max
threshold = ( predictions.max() + predictions.min()) / 2
threshold
```

```python
binary_logic = lambda val: 1 if val > threshold else 0
p = [ binary_logic(prediction) for prediction in predictions]
p[:5]
```

```python
y_test_encoded[:5]
```

```python
y_test_encoded[:15]
```

```python
from sklearn.metrics import accuracy_score
ascore = accuracy_score(p, y_test_encoded)
```

```python
print('Linear regression accuracy was ' + str(ascore*100) + '%')
```
