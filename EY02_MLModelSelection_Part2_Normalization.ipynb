```python
# Min Max -> (data-min) / (max - min)
# advantage -> all values become between 0 to 1 which is a standard scale-> math rules are preserved!
# disadvantage -> outliers (if too large or too small), our values of min and max will get impacted,
# hence disturbing the entire scaling, and make normal values very very tiny!

# Z-Score -> 2 things-> it scales your data such that mean value becomes 0, and all the other value get
# scaled with respect to standard deviation
# formula -> (data - mean)/standard_deviation
# advantage-> outliers can be detected and treated (dropped, adjusted, ignored...)
# disadvantage-> (-inf to +inf)
# solution-> we only consider a smaller scale of (-3 to +3)
#     anything more than 3 times of standard deviation is CONSIDERED outlier!
#     anything within this limit is considered a normal point

# Central Limit Theorem (Please find screenshot in github)
```

```python
URL = 'https://raw.githubusercontent.com/a-forty-two/EY8Apr2024-AI-Batch2/main/data.csv'
data = pd.read_csv(URL, header=0)
data.set_index('id', inplace=True)
```

```python
data.columns
```

```python
y = data['diagnosis']
x = data.iloc[:, 1:-1]
```

```python
x.columns
```

```python
# NORMALIZATION RULES
# First we divide our data into training and scoring
# we calculate normalization parameters ONLY for the training data [ASSUMPTIONS]
# we apply the same assumptions to scoring dataset as well

# for example-> if we were MinMax, then min max will be calculated from training dataset & applied
# to scoring dataset

# if we were using Z-Score, then mean and std will be calculated from training dataset & applied
# to scoring dataset

```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.20)
```

```python
# Z-score-> mu (mean) and sigma (std)
xtrain.describe()
```

```python
xtrain.describe().T
```

```python
mu = xtrain.describe().T['mean']
sigma = xtrain.describe().T['std']
```

```python
mu
```

```python
#normalize the data for training and testing
xtrain_norm = (xtrain - mu) / sigma
xtest_norm = (xtest - mu) / sigma
```

```python
xtrain_norm.head()
```

```python
xtrain_norm[xtrain_norm > 3]
```

```python
xtrain_norm.max()
```

```python
# ML Engineer-> 3 options
# 1. Dropped them [BEST]
# 2. Treated them
#       a. Replace them -> Min (> -3) or Max (< +3) [BEST]
#       b. Replace them with 3 or -3
# 3. IGNORE THEM [BEST] -> a statistics analyst will never take this approach
```
