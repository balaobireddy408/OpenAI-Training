```python
# 3 most popular libraries-> NLTK, Spacy, Textacy 

import spacy 
```

```python
dictionary = spacy.load('en_core_web_sm')
```

```python
text = """Mary had a little lamb, and little lamb said 'bye bye i am going home'. 
This sounds very strange, because lamb was planning to go on a dinner with the goat. 
Lamb started watching a movie and absolutely hated it."""

```

```python
# Document -> pages -> paragraphs 
# SENTENCES -> WORDS -> TOKENS
# MULTIPLE  worlds can also form a token-> 'The Himalayas', 'didn't'

nlp = dictionary(text)
print(dir(nlp))
```

```python
nlp.vector
```

```python
# tokens inside the sentences 
for i in range(len(nlp)):
    print(' [ ' + str(nlp[i]), end=' ], ')
    
    
```

```python
for i in range(len(nlp)):
    print(nlp[i], end=' = ')
    print(nlp[i].vector)
```

```python

for i in range(len(nlp)):
    print('Grammar')
    print(nlp[i], end=' is a ')
    print(nlp[i].pos_)
    print('Tag = ', end = '')
    print(nlp[i].tag_)
```

```python
# Objective = collect all nouns from the data 
all_nouns = []
for i in range ( len (nlp)):
    if(nlp[i].pos_ == 'NOUN' or nlp[i].pos_=='PROPN'):
        all_nouns.append(nlp[i])
    
```

```python
# Objective = collect all nouns from the data 
all_nouns = []
for i in range ( len (nlp)):
    if(nlp[i].pos_ == 'NOUN' or nlp[i].pos_=='PROPN'):
        all_nouns.append(str(nlp[i]).lower())
```

```python
# unique nouns 
all_nouns_unique = set(all_nouns)
```
