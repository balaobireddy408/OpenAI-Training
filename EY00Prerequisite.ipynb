Dictionary -> SET OF LISTS [DATA SCIENCE ONLY]

key value pairs-> key is column name, values are column data

```python
mydata = {
    'Name' : ['John', 'Pikachu', 'Tom', 'Harry'],
    'Age' : [10, 20, 30, 40]
}
print(mydata)
```

but besides data science, a dictionary could have stored anything as key and anything as value!


```python
mydict = {
    0: 'hello world',
    '1.0' : mydata,
    False : myset,
    -1 : profittuple
}
print(mydict)
```

```python
mydict[-1]
```

```python
# Dictionary where values have the same length can be converted into a data frame
# data frame = ??

import pandas as pd

data = pd.DataFrame(mydata)
data
```

```python


# 2 very important methods-
# a. selection by name -> loc
# b. selection by index -> iloc

#format -> dataframe.iloc [ row_selection, column_selection]
data.iloc[: , :] # all the rows, and all columns
```

```python
# all values but only from the last column
data.iloc[:, [-1]] # pass indexes either as a list or as a range
```

```python
# otherwise, we get returned a numpy array and not a dataframe which can mean future errors
data.iloc[:, -1] # observe that column name is missing
```

```python
# use loc when you know the column names!
data.loc[:, ['Age']]
```

```python
# row selection is similar
# we want the last 2 rows only for the age column
# using loc
data.loc[ -2:, ['Age']] # this approach can be buggy -> best to avoid
```

```python
# using iloc
data.iloc[-2:, -1:]
```
